```python
import json
import requests
import pandas as pd
import matplotlib.pyplot as plt
import boto3
from io import StringIO

BUCKET_NAME = "lab-s3-2026-12345"   
START_DATE = "20230101"
END_DATE = "20231231"
```

```python
def fetch_nbu_rates(valcode, start_date, end_date):
    url = (
        "https://bank.gov.ua/NBU_Exchange/exchange_site"
        f"?start={start_date}&end={end_date}"
        f"&valcode={valcode}&sort=exchangedate&order=asc&json"
    )
    response = requests.get(url, timeout=30)
    response.raise_for_status()
    return response.json()

usd_json = fetch_nbu_rates("USD", START_DATE, END_DATE)
eur_json = fetch_nbu_rates("EUR", START_DATE, END_DATE)

print(f"USD records: {len(usd_json)}")
print(f"EUR records: {len(eur_json)}")
print("Sample USD row:", usd_json[0])
```

```python
with open("usd_2022.json", "w", encoding="utf-8") as f:
    json.dump(usd_json, f, ensure_ascii=False, indent=2)

with open("eur_2022.json", "w", encoding="utf-8") as f:
    json.dump(eur_json, f, ensure_ascii=False, indent=2)

print("JSON files saved.")
```

```python
def json_to_dataframe(data):
    df = pd.DataFrame(data)
    df = df[["cc", "exchangedate", "rate"]].copy()
    df.columns = ["currency", "date", "rate"]
    return df

usd_df = json_to_dataframe(usd_json)
eur_df = json_to_dataframe(eur_json)

usd_df.to_csv("usd_2022.csv", index=False, encoding="utf-8")
eur_df.to_csv("eur_2022.csv", index=False, encoding="utf-8")

print("CSV files created.")
usd_df.head()
```

```python
s3 = boto3.client("s3")

s3.upload_file("usd_2022.csv", BUCKET_NAME, "rates/usd_2022.csv")
s3.upload_file("eur_2022.csv", BUCKET_NAME, "rates/eur_2022.csv")

print(f"Uploaded CSV files to s3://{BUCKET_NAME}/rates/")
```

```python
def read_csv_from_s3(bucket, key):
    obj = s3.get_object(Bucket=bucket, Key=key)
    content = obj["Body"].read().decode("utf-8")
    return pd.read_csv(StringIO(content))

usd_s3 = read_csv_from_s3(BUCKET_NAME, "rates/usd_2022.csv")
eur_s3 = read_csv_from_s3(BUCKET_NAME, "rates/eur_2022.csv")

usd_s3.head(), eur_s3.head()
```

```python
usd_s3["date"] = pd.to_datetime(usd_s3["date"], format="%d.%m.%Y")
eur_s3["date"] = pd.to_datetime(eur_s3["date"], format="%d.%m.%Y")

usd_s3 = usd_s3.sort_values("date")
eur_s3 = eur_s3.sort_values("date")

print(usd_s3.dtypes)
```

```python
plt.figure(figsize=(14, 6))
plt.plot(usd_s3["date"], usd_s3["rate"], label="USD/UAH")
plt.plot(eur_s3["date"], eur_s3["rate"], label="EUR/UAH")

plt.title("Official UAH exchange rate to USD and EUR in 2022")
plt.xlabel("Date")
plt.ylabel("Exchange rate")
plt.legend()
plt.grid(True)
plt.tight_layout()

plt.savefig("uah_usd_eur_2022.png", dpi=150)
plt.show()
```
